# Highly divisible triangular number

Problem 12
The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
from datetime import datetime
start_time = datetime.now()

def factors(n):
    factors = []
    for i in range(1, n + 1):
        if n % i == 0:
            factors.append(i)
    return factors

n = 2507680
i = 2239

while True:
    if len(factors(n)) == 500:
        break
    else:
        n = n + (i +1)
        i += 1

print(n)

end_time = datetime.now()
time_diff = (end_time - start_time)
execution_time = time_diff.total_seconds() * 1000

print(execution_time)
```

```python
import math

from datetime import datetime
start_time = datetime.now()

def sieve_erat(n):
    A = [1 for i in range(n+1)]
    A[0] = 0
    A[1] = 0
    for i in range(4, n+1, 2):
        A[i] = 0
    for i in range(6, n+1, 3):
        A[i] = 0

    for i in range(5, math.ceil(math.sqrt(n)), 6):
        if A[i]:
            j = i * 2
            while j <= n:
                A[j] = 0
                j += i

        if A[i +2]:
            j = (i+2) * 2
            while j <= n:
                A[j] = 0
                j += (i+2)

    P = [i for i in range(n) if A[i] == 1]

    return P

def dividers(k, factors):

    unique_factors_of_k = []

    for i in factors:
        if i > k == 0:
            break
        if k % i == 0:
            unique_factors_of_k.append(i)

    unique_factors_of_k.append(k)

    num_dividers = 1

    for i in unique_factors_of_k:
        if i == 1:
            continue
        count = 1
        while k % i == 0:
            count += 1
            k = k / i

        num_dividers *= count

    return num_dividers

p = 100
prime_factors_under_p = sieve_erat(int(p + 1))
prime_factors_under_p.insert(0, 1)

max_num_divisors = 500

triangle_number = 1
i = 1

while True:
    if dividers(triangle_number, prime_factors_under_p) > max_num_divisors:
        print(triangle_number)
        break
    else:
        triangle_number = triangle_number + (i + 1)
        i += 1


end_time = datetime.now()
time_diff = (end_time - start_time)
execution_time = time_diff.total_seconds() * 1000

print(execution_time)
```
